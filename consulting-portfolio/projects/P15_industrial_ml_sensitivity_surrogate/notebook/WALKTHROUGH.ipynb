# Walkthrough (sin código)

**Proyecto:** P15 — ML sensibilidad + surrogate modeling

Walkthrough del flujo de sensibilidad y surrogate model (sin código ni datos)


## 1) Contexto y objetivo

- Un optimizador industrial produce KPIs (costos, downtime, horas fuera de operación, etc.).
- El equipo necesitaba aproximar esos KPIs *sin* correr el optimizador para cada iteración.


## 2) Pipeline de decisión

- Ingesta de escenarios → limpieza → definición de features/targets.
- Entrenamiento de regresores + evaluación.
- Cálculo de sensibilidades y ranking de variables.
- Export de reporte (notebook/HTML).


## 3) Contratos de datos (forma, no contenido)

- Inputs: parámetros numéricos/categóricos (configuración).
- Outputs: KPIs agregados por escenario (tabular).
- Constraints: rangos válidos, nulos, tipos.


## 4) Testing y observabilidad

- Tests de esquema (tipos, rangos, nulos).
- Tests de leakage (features prohibidas).
- Registro de métricas y drift checks.


## 5) Riesgos + mitigaciones

- Riesgo de conclusiones erróneas → baselines + validación robusta.
- Riesgo de exposición de datos → repo público solo con documentación.
