# Walkthrough (sin código)

**Proyecto:** P16 — Multi‑stage optimisation

Walkthrough del driver multi‑etapa (relax → fix → final) y del diagnóstico de infeasibilidad


## 1) Contexto y objetivo

- Optimización industrial con miles de binarios operacionales y restricciones temporales.
- Necesidad: estabilidad + tiempo de cómputo predecible.


## 2) Pipeline de decisión

- Discovery de componentes binarios.
- Solve relajado para obtener perfiles.
- Fijación de perfiles/segmentos.
- Solve final + auditoría.
- Si infeasible: export → feasRelaxS → slacks → reporte.


## 3) Contratos de datos (forma, no contenido)

- Perfiles on/off: series 0/1 por tiempo.
- SegmentFix: índice de segmento por tiempo (eficiencia variable).
- Reporte slacks: dict por componente/operación.


## 4) Testing y observabilidad

- Tests: compatibilidad de perfiles, dominios y shapes.
- Observabilidad: logs por etapa, tiempos, estado solver, tamaño modelo.


## 5) Riesgos + mitigaciones

- Mitigaciones: fallback iterativo, tolerancias, diagnóstico guiado para ajustar restricciones.
- Control de fuga: no publicar datos/MPS.
