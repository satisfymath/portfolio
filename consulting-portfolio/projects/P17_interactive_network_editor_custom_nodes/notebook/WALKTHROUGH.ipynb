# Walkthrough (sin código)

**Proyecto:** P17 — Network editor custom nodes

Walkthrough del editor de redes (UI + estado + export)


## 1) Contexto y objetivo

- Necesidad de diseñar relaciones/nodos visualmente con consistencia de layout.
- Requerimiento: export y control de UI sin backend.


## 2) Pipeline de decisión

- Capa de renderer (canvas).
- Nodos custom por tipo (estilos/handles).
- Toolbar para interacción.
- Estado serializable para persistencia.


## 3) Contratos de datos (forma, no contenido)

- Node schema: {id, type, data, position}.
- Handle schema: in/out con metadata de ángulo/side.
- Export: PNG del viewport.


## 4) Testing y observabilidad

- Tests: render smoke, helpers de normalización, serialización idempotente.
- Telemetría: tiempos de export, tamaño de grafo (opcional).


## 5) Riesgos + mitigaciones

- Riesgo de UI inconsistente → contratos + normalización.
- Riesgo de datos sensibles en export → advertencias/controles en producto final.
